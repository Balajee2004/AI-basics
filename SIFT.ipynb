# What is SIFT Algorithm? 

The SIFT (Scale-Invariant Feature Transform) algorithm is a computer vision technique used for detecting and describing local features in images. It was invented by David Lowe in 1999Â². The algorithm identifies key points in an image that are invariant to scale and rotation, and then generates descriptors for these key points which can be used for image matching, object recognition, and other image processing tasks.

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#reading images
img1 = cv2.imread("C:/Users/MR.H.BALAJEE/Downloads/india gate1.jpeg")
img1 = cv2.resize(img1,(250,250))
img2 = cv2.imread("C:/Users/MR.H.BALAJEE/Downloads/india gate2.jpeg")
img2 = cv2.resize(img2,(250,250))

#Convert to gray scale
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY)
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY)

#sift
sift = cv2.SIFT_create()

keypoint_1, descriptors_1 = sift.detectAndCompute(img1,None)
keypoint_2, descriptors_2 = sift.detectAndCompute(img2,None)
```
