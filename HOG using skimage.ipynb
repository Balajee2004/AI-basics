# Importing Required Libraries

```python
from skimage.io import imread
from skimage.transform import resize
from skimage.feature import hog
import matplotlib.pyplot as plt
```

# Reading the Image


```python
img = imread("C:/Users/MR.H.BALAJEE/Downloads/german shepherd.jpeg")
plt.axis("off")
plt.imshow(img)
print(img.shape)
```

# Creating HOG Features

```python
fd, hog_image = hog(img, orientations = 10, pixels_per_cell = (2, 2), cells_per_block = (4, 4), visualize = True, channel_axis = 2)
#Show Image
print(fd.shape)
print(hog_image.shape)
plt.axis("off")
plt.imshow(hog_image, cmap = 'hot')
plt.show()
```
