# Importing Required Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('prices_to_predict.csv', low_memory=False)
```

# Numpy Functions

# Array Creation

```python
# np.array(): Create an array from a list
arr = np.array([1, 2, 3, 4, 5])
print("Array:", arr)

# np.zeros(): Create an array of zeros
zeros = np.zeros(5)
print("Zeros:", zeros)

# np.ones(): Create an array of ones
ones = np.ones(5)
print("Ones:", ones)

# np.arange(): Create an array with a range of values
range_arr = np.arange(10)
print("Range Array:", range_arr)

# np.linspace(): Create an array with evenly spaced values
linspace_arr = np.linspace(0, 1, 5)
print("Linspace Array:", linspace_arr)

# np.eye(): Create an identity matrix
identity_matrix = np.eye(3)
print("Identity Matrix:\n", identity_matrix)

# np.random.rand(): Create an array with random values
random_arr = np.random.rand(5)
print("Random Array:", random_arr)
```

## Array Operations

```python
# np.add(): Add two arrays
sum_arr = np.add(arr, ones)
print("Sum Array:", sum_arr)

# np.subtract(): Subtract one array from another
diff_arr = np.subtract(arr, ones)
print("Difference Array:", diff_arr)

# np.multiply(): Multiply two arrays
prod_arr = np.multiply(arr, ones)
print("Product Array:", prod_arr)

# np.divide(): Divide one array by another
quot_arr = np.divide(arr, ones)
print("Quotient Array:", quot_arr)

# np.dot(): Dot product of two arrays
dot_product = np.dot(arr, arr)
print("Dot Product:", dot_product)

# np.sqrt(): Square root of each element in the array
sqrt_arr = np.sqrt(arr)
print("Square Root Array:", sqrt_arr)

# np.power(): Raise each element in the array to a power
power_arr = np.power(arr, 2)
print("Power Array:", power_arr)
```

## Statistical Operations

```python
# np.mean(): Mean of the array
mean = np.mean(arr)
print("Mean:", mean)

# np.median(): Median of the array
median = np.median(arr)
print("Median:", median)

# np.std(): Standard deviation of the array
std = np.std(arr)
print("Standard Deviation:", std)

# np.var(): Variance of the array
var = np.var(arr)
print("Variance:", var)

# np.sum(): Sum of the array
total = np.sum(arr)
print("Sum:", total)

# np.min() and np.max(): Minimum and maximum of the array
min_val = np.min(arr)
max_val = np.max(arr)
print("Min:", min_val, "Max:", max_val)

# np.argmin() and np.argmax(): Indices of the minimum and maximum values
min_idx = np.argmin(arr)
max_idx = np.argmax(arr)
print("Min Index:", min_idx, "Max Index:", max_idx)
```

## Array Manipulation

```python
# np.reshape(): Reshape the array
reshaped = arr.reshape(1, 5)
print("Reshaped Array:\n", reshaped)

# np.transpose(): Transpose the array
transposed = reshaped.transpose()
print("Transposed Array:\n", transposed)

# np.concatenate(): Concatenate two arrays
concatenated = np.concatenate((arr, arr))
print("Concatenated Array:", concatenated)

# np.vstack() and np.hstack(): Stack arrays vertically and horizontally
vstacked = np.vstack((arr, arr))
hstacked = np.hstack((arr, arr))
print("Vertically Stacked Array:\n", vstacked)
print("Horizontally Stacked Array:\n", hstacked)

# np.split(): Split the array into multiple sub-arrays
split_arr = np.split(arr, 5)
print("Split Arrays:", split_arr)

# np.where(): Elements satisfying the condition
condition = np.where(arr > 3)
print("Elements greater than 3:", condition)

# np.unique(): Unique elements in the array
unique_elements = np.unique(arr)
print("Unique Elements:", unique_elements)
```

## Data Cleaning and Preparation

```python
# Load the dataset
df = pd.read_csv('prices_to_predict.csv', low_memory=False)

# Print the column names to identify the correct ones
print(df.columns)

# Check if the column 'T_degC' exists before dropping it
if 'T_degC' in df.columns:
    # df.drop(): Drop rows or columns
    df_dropped = df.drop(columns=['T_degC'])  # Replace 'T_degC' with the actual column name to drop
    print("DataFrame after dropping column 'T_degC':\n", df_dropped)
else:
    print("Column 'T_degC' does not exist in the dataset.")

# df.dropna() and df.fillna(): Drop or fill missing values
df_no_na = df.dropna()
df_filled_na = df.fillna(0)
print("DataFrame without NA:\n", df_no_na)
print("DataFrame with NA filled:\n", df_filled_na)

# df.replace(): Replace values
df_replaced = df.replace(1, 100)
print("DataFrame with values replaced:\n", df_replaced)

# df.duplicated() and df.drop_duplicates(): Identify and drop duplicate rows
duplicates = df.duplicated()
df_no_duplicates = df.drop_duplicates()
print("Duplicates:\n", duplicates)
print("DataFrame without duplicates:\n", df_no_duplicates)

# df.astype(): Change the data type of a column
# Replace 'T_degC' with the actual column name you want to change the data type
if 'T_degC' in df.columns:
    df['T_degC'] = df['T_degC'].astype(float)
    print("DataFrame with column 'T_degC' as float:\n", df)
else:
    print("Column 'T_degC' does not exist in the dataset.")
```

## Data Transformation

```python
# Load the dataset
df = pd.read_csv('prices_to_predict.csv', low_memory=False)

# Print the column names to identify the correct ones
print(df.columns)

# Check if the columns 'T_degC' and 'Depthm' exist before applying operations
if 'T_degC' in df.columns and 'Depthm' in df.columns:
    # df.apply() and df.map(): Apply functions to columns or DataFrame
    df_applied = df['T_degC'].apply(np.sqrt)
    df_mapped = df['T_degC'].map(lambda x: x + 1)
    print("Applied sqrt to column 'T_degC':\n", df_applied)
    print("Mapped lambda to column 'T_degC':\n", df_mapped)

    # df.groupby() and df.agg(): Group by and aggregate data
    grouped = df.groupby('Depthm').agg({'T_degC': 'mean'})
    print("Grouped by column 'Depthm' and mean of column 'T_degC':\n", grouped)

    # df.merge() and df.join(): Merge or join DataFrames
    merged = df.merge(df, on='T_degC')
    print("Merged DataFrame:\n", merged)

    # df.concat(): Concatenate DataFrames
    concatenated = pd.concat([df, df])
    print("Concatenated DataFrame:\n", concatenated)
else:
    print("Columns 'T_degC' and/or 'Depthm' do not exist in the dataset.")
```
