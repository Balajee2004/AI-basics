# Using Numpy Functions

## Array Creation

```python
# np.array(): Create an array from a list
arr = np.array([1, 2, 3, 4, 5])
print("Array:", arr)

# np.zeros(): Create an array of zeros
zeros = np.zeros(5)
print("Zeros:", zeros)

# np.ones(): Create an array of ones
ones = np.ones(5)
print("Ones:", ones)

# np.arange(): Create an array with a range of values
range_arr = np.arange(10)
print("Range Array:", range_arr)

# np.linspace(): Create an array with evenly spaced values
linspace_arr = np.linspace(0, 1, 5)
print("Linspace Array:", linspace_arr)

# np.eye(): Create an identity matrix
identity_matrix = np.eye(3)
print("Identity Matrix:\n", identity_matrix)

# np.random.rand(): Create an array with random values
random_arr = np.random.rand(5)
print("Random Array:", random_arr)
```

## Array Operations

```python
# np.add(): Add two arrays
sum_arr = np.add(arr, ones)
print("Sum Array:", sum_arr)

# np.subtract(): Subtract one array from another
diff_arr = np.subtract(arr, ones)
print("Difference Array:", diff_arr)

# np.multiply(): Multiply two arrays
prod_arr = np.multiply(arr, ones)
print("Product Array:", prod_arr)

# np.divide(): Divide one array by another
quot_arr = np.divide(arr, ones)
print("Quotient Array:", quot_arr)

# np.dot(): Dot product of two arrays
dot_product = np.dot(arr, arr)
print("Dot Product:", dot_product)

# np.sqrt(): Square root of each element in the array
sqrt_arr = np.sqrt(arr)
print("Square Root Array:", sqrt_arr)

# np.power(): Raise each element in the array to a power
power_arr = np.power(arr, 2)
print("Power Array:", power_arr)
```

## Statistical Operations

```python
# np.mean(): Mean of the array
mean = np.mean(arr)
print("Mean:", mean)

# np.median(): Median of the array
median = np.median(arr)
print("Median:", median)

# np.std(): Standard deviation of the array
std = np.std(arr)
print("Standard Deviation:", std)

# np.var(): Variance of the array
var = np.var(arr)
print("Variance:", var)

# np.sum(): Sum of the array
total = np.sum(arr)
print("Sum:", total)

# np.min() and np.max(): Minimum and maximum of the array
min_val = np.min(arr)
max_val = np.max(arr)
print("Min:", min_val, "Max:", max_val)

# np.argmin() and np.argmax(): Indices of the minimum and maximum values
min_idx = np.argmin(arr)
max_idx = np.argmax(arr)
print("Min Index:", min_idx, "Max Index:", max_idx)
```

## Array Manipulation

```python
# np.reshape(): Reshape the array
reshaped = arr.reshape(1, 5)
print("Reshaped Array:\n", reshaped)

# np.transpose(): Transpose the array
transposed = reshaped.transpose()
print("Transposed Array:\n", transposed)

# np.concatenate(): Concatenate two arrays
concatenated = np.concatenate((arr, arr))
print("Concatenated Array:", concatenated)

# np.vstack() and np.hstack(): Stack arrays vertically and horizontally
vstacked = np.vstack((arr, arr))
hstacked = np.hstack((arr, arr))
print("Vertically Stacked Array:\n", vstacked)
print("Horizontally Stacked Array:\n", hstacked)

# np.split(): Split the array into multiple sub-arrays
split_arr = np.split(arr, 5)
print("Split Arrays:", split_arr)

# np.where(): Elements satisfying the condition
condition = np.where(arr > 3)
print("Elements greater than 3:", condition)

# np.unique(): Unique elements in the array
unique_elements = np.unique(arr)
print("Unique Elements:", unique_elements)
```

## Data Cleaning and Preparation

```python
# Load the dataset
df = pd.read_csv('student.csv')

# Display the first few rows of the DataFrame
print("Original DataFrame:\n", df.head())

# 1. df.drop(): Drop the 'ParentHighestLevelEducation' column (example column name)
df_dropped = df.drop(columns=['Course'])
print("\nDataFrame after dropping 'ParentHighestLevelEducation' column:\n", df_dropped.head())

# 2. df.dropna(): Drop rows with any missing values
df_no_na = df.dropna()
print("\nDataFrame after dropping rows with missing values:\n", df_no_na.head())

# 3. df.fillna(): Fill missing values with a specified value (e.g., 0)
df_filled_na = df.fillna(0)
print("\nDataFrame after filling missing values with 0:\n", df_filled_na.head())

# 4. df.replace(): Replace values (e.g., replace 'Yes' in 'FirstGenStudent' column with 'Y')
df_replaced = df.replace({'FirstGenStudent': {'Yes': 'Y', 'No': 'N'}})
print("\nDataFrame after replacing values in 'FirstGenStudent' column:\n", df_replaced.head())

# 5. df.duplicated(): Check for duplicate rows
# (Adding a duplicate row for demonstration)
df.loc[len(df)] = df.iloc[0]
duplicates = df.duplicated()
print("\nDuplicate rows:\n", duplicates)

# 6. df.drop_duplicates(): Drop duplicate rows
df_no_duplicates = df.drop_duplicates()
print("\nDataFrame after dropping duplicate rows:\n", df_no_duplicates.head())

# 7. df.astype(): Change the data type of a column (e.g., 'Retention' to float)
df['course'] = df['Course'].astype(float)
print("\nDataFrame with 'Retention' as float:\n", df.head())
```

## Data Transformation

```python
data = {
    'A': [1, 2, 3],
    'B': [4, 5, 6],
    'C': [7, 8, 9]
}

df = pd.DataFrame(data)

# df.apply()
df_apply = df.apply(lambda x: x * 2)
print("df.apply():\n", df_apply)

# df.map()
df_map = df['A'].map(lambda x: x * 2)  # This will still work as expected
print("\ndf.map():\n", df_map)

# df.groupby()
df_groupby = df.groupby('C').sum()
print("\ndf.groupby():\n", df_groupby)

# df.agg()
df_agg = df.agg({'A': 'sum', 'B': 'mean'})
print("\ndf.agg():\n", df_agg)

# df.merge()
df2 = pd.DataFrame({'C': [7, 8, 9], 'D': [10, 11, 12]})
df_merge = pd.merge(df, df2, on='C')
print("\ndf.merge():\n", df_merge)

# df.join()
df3 = pd.DataFrame({'C': [7, 8, 9], 'E': [13, 14, 15]})
df_join = df.join(df3.set_index('C'), on='C')
print("\ndf.join():\n", df_join)

# df.concat()
df_concat = pd.concat([df, df2], axis=1)
print("\ndf.concat():\n", df_concat)
```

## Statistical Operations

```python
# Given DataFrame
data = {
    'A': [1, 2, 3],
    'B': [4, 5, 6],
    'C': [7, 8, 9]
}

df = pd.DataFrame(data)

# df.mean() - calculate mean
mean_values = df.mean()
print("Mean values:\n", mean_values)

# df.median() - calculate median
median_values = df.median()
print("\nMedian values:\n", median_values)

# df.std() - calculate standard deviation
std_values = df.std()
print("\nStandard deviation values:\n", std_values)

# df.var() - calculate variance
variance_values = df.var()
print("\nVariance values:\n", variance_values)

# df.sum() - calculate sum
sum_values = df.sum()
print("\nSum values:\n", sum_values)

# df.min() - find minimum values
min_values = df.min()
print("\nMinimum values:\n", min_values)

# df.max() - find maximum values
max_values = df.max()
print("\nMaximum values:\n", max_values)

# df.corr() - calculate correlation matrix
correlation_matrix = df.corr()
print("\nCorrelation matrix:\n", correlation_matrix)

# df.cov() - calculate covariance matrix
covariance_matrix = df.cov()
print("\nCovariance matrix:\n", covariance_matrix)
```
